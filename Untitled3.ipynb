```python

class Node:
    def __init__(self, value): #ссылка на узел и значение узла 
        self.value = value
        self.next = None


# версия 1
class LinkedListV1:  #базовый списко
    def __init__(self):
        self.head = None #указатель на 1 узел в списке

    def append(self, value):  #метод "добавить в конец"
        if not self.head:
            self.head = Node(value) #если список пустой создаем новый узел и делаем его головой
            return
        cur = self.head  #ТЕКУЩИЙ
        while cur.next:  #пока у текущего узла есть следующий
            cur = cur.next  #перемещаемся к след узлу и так до последнего
        cur.next = Node(value)

    def print_list(self):
        cur = self.head
        while cur:
            print(cur.value, end=" -> ")   #пока есть узел, печатаем его значение
            cur = cur.next
        print("None")


#версия 2
class LinkedListV2(LinkedListV1):  #этот класс получает все свойства первого
    def prepend(self, value):  # метод "добавитбь в начало"
        node = Node(value)
        node.next = self.head  #след узел явл текущей головой списка
        self.head = node  #делаем новый узел головой спискка


#версия 3
class LinkedListV3(LinkedListV2):
    def delete(self, value):
        cur = self.head
        prev = None  #предыдущий узел

        while cur:
            if cur.value == value:
                if prev:
                    prev.next = cur.next  #следующий теперь следующий после удаляемого
                else:
                    self.head = cur.next   #удаляем первый элемент
                return
            prev = cur  #текущий становится предыдущим
            cur = cur.next


#версия 4
class LinkedListV4(LinkedListV3):
    def find(self, value):
        cur = self.head
        while cur:
            if cur.value == value:  #если нашли нужный узел
                return True
            cur = cur.next
        return False


#версия 5
class LinkedListV5(LinkedListV4):
    def length(self):
        count = 0
        cur = self.head
        while cur:
            count += 1
            cur = cur.next
        return count


#версия 6
class LinkedListV6(LinkedListV5):
    def get(self, index):  #по индексу
        cur = self.head
        i = 0
        while cur:
            if i == index:
                return cur.value
            cur = cur.next
            i += 1
        raise IndexError("Index out of range")

    # 2. reverse
    def reverse(self):
        prev = None
        cur = self.head
        while cur:
            nxt = cur.next
            cur.next = prev
            prev = cur
            cur = nxt
        self.head = prev

    # 3. sort
    def sort(self):
        if not self.head:
            return

        swapped = True  #флаг есть ли обмены
        while swapped:
            swapped = False
            cur = self.head
            while cur.next:
                if cur.value > cur.next.value:  #если текущ больше след, то меняем местами
                    cur.value, cur.next.value = cur.next.value, cur.value
                    swapped = True
                cur = cur.next

    # 4. индивидуальное задание вариант 11
    def individual_task_11(self):
        """Удалить все нечетные числа из списка"""
        cur = self.head
        prev = None

        while cur:
            if cur.value % 2 != 0:
                if prev:
                    prev.next = cur.next  #как в 3 методе
                else:
                    self.head = cur.next
                cur = cur.next
                continue

            prev = cur
            cur = cur.next


# Тестирование
if __name__ == "__main__":
    lst = LinkedListV6()

    # Заполнение списка
    nums = [1, 5, 2, 9, 4, 7, 6, 8, 3]
    for n in nums:
        lst.append(n)

    print("Исходный список:")
    lst.print_list()

    # Reverse
    lst.reverse()
    print("После reverse:")
    lst.print_list()

    # Sort
    lst.sort()
    print("После sort:")
    lst.print_list()

    # Индивидуальное задание №11
    lst.individual_task_11()
    print("После удаления нечётных чисел:")
    lst.print_list()

    # Дополнительно
    print("Длина списка:", lst.length())
    print("Поиск числа 4:", lst.find(4))
 
```
